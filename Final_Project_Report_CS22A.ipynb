```python
# Calculate the pearson correlation and p-value using scipy stats.
from scipy import stats
correl, p_val = stats.pearsonr(x, y)
print('pearson correlation =', round(correl, 2))
print('p_value =', p_val)


```

## Question 2 : Is there a difference between the cost of small dog breeds and the overall population of dog breeds? At alpha level = 0.05

**Null hypothesis** : Null mean (mean average cost of small dog breed) = mean cost of population of all dog breeds
* This means there is no difference between the mean average cost of small dog breeds and the overall population

**Alternative hypothesis** : Null mean (mean average cost of small dog breed) ≠ mean cost of population of all dog breeds
* This mean there is a difference between the mean average cost of small dog breeds and the overall population

Here, we chose a sample size of 30 because according to the Central Limit Theorem (CLT), the distribution of the sample mean is normal

```python
# T-critical values
sample_size = 30
df = sample_size - 1
t_neg = -2.045
t_pos = +2.045

##def find_t_statistic(SE, mean1, mean2, sample_size):
 ##t_stats = (mean1 - mean2) / (SE / math.sqrt(sample_size))

 ##return t_stats
```

* Because we have a sample size of 30, after going through the t-table, we got the t_critcal values -2.045 and 2.045 in a two tailed sample test which serves as a boundary to whether we reject or fail to reject our null hypothesis.

**Step 1**: Find the mean of the sample group and the sample.

```python
## Small dog’s mean
import pandas as pd
dog_df = pd.read_csv('https://raw.githubusercontent.com/csbfx/cs22a/main/AKC_dogs_rankings.csv')
# Small breed average cost
sb_cost = dog_df[dog_df['Size'] == 'small']['Avg_price']
# Small breed average cost mean
sbc_m = sb_cost.mean()
# Small breed average cost standard deviation
sbc_std = sb_cost.std()
print('The mean of average small dog price is:', round(sbc_m, 2))

# Overall dog mean
dogc_m = dog_df['Avg_price'].mean()

print('The mean of average dog price is', round(dogc_m, 2))
```

**Step 2** : Find the Standard Error.

```python
## Find the SE:
import math
dog_df['squared_deviation'] = (sbc_m - dogc_m)**2
sample_variance = sum(dog_df["squared_deviation"]) / sample_size
SE = math.sqrt(sample_variance)

print(f"SE: {round(SE,2)}")
```

* A standard error of 57.89, means that sample mean of small dogs is likely to vary around 57.89 from the true population.

**Step 3**: Calculate the T-statistic by applying calculations from step 1 and 2 into the T-statistic formula:

** t = (X‾ - μ0) / (s / √n) **

```python
# Calculate T-statistic:
import math

# t = (x_bar - m) / (s / math.sqrt(n))

t = (sbc_m - dogc_m) / (SE / math.sqrt(sample_size - 1))
print('t statistic=', round(t,2))

```

**Step 4** : LAST, check to see if the t-statistic is in the interval of the t-critical values to see if we have evidence to reject the null.

```python
if t < t_neg or t > t_pos:
	print('reject null hypothesis')

```

**Final Analysis** :

In the previous cell, we got the result of the t-statistic is which *rejected the null hypothesis.*

Since we had a t-negative and t-positive at -2.045 and 2.045, the t-statistic of -2.25, which is smaller than the t-negative, means that we can reject the null hypothesis. Therefore this means there is a difference between the mean average cost of small dog breeds and the overall population.
